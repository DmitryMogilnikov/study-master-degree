```python
import pandas as pd
```

```python
path = 'Nizhnevartovsk_weather_2017-2022.csv'

df = pd.read_csv(path, sep=';',index_col=False)
```

```python
df
```

```python
df['Местное время в Нижневартовске / им. В. И. Муравленко (аэропорт)'] = pd.to_datetime(df['Местное время в Нижневартовске / им. В. И. Муравленко (аэропорт)'])
df
```

```python
min_temp = df.groupby([df['Местное время в Нижневартовске / им. В. И. Муравленко (аэропорт)'].dt.date]).agg(min_temperature = pd.NamedAgg(column = 'Tn', aggfunc = 'min'))
min_temp.to_csv('Nizhnevartovsk_min_temperature.csv')
```
